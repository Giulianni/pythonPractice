```python
import pandas as pd
import numpy as np 
from datetime import datetime 
import os
```

```python
def antinaEspecialLoanCargar(archivoActual, fechaDelDia, archivoObtenidoFormatoLoan):

    archivoLectura = archivoActual
    fechaDelDia = fechaDelDia
    escrituraArchivo = archivoObtenidoFormatoLoan

    df = pd.read_excel(archivoLectura)

    
    #Verificar columnas-----------------------------------------------------------------------
    df.columns = df.columns.astype(str)
    columnas_df = df.columns  

    print(columnas_df)

    columnas_validas_df = ['DNI', fechaDelDia] #Ingreso columnas que necesito para manejarlas y solucionar mi problema
    
    for i in columnas_df:
        if i in  columnas_validas_df:
            columnas_validas_df.remove(i)
        if not columnas_validas_df:
            print('Columnas válidas del Dataframe anterior')
    else: 
            print(f'Columnas faltantes del Dataframe anterior: {columnas_validas_df}')

            #df.columns #Para fijarme los nombres de las columnas 

    #Utils --------------------------------------------------------------------------------
    columnasVacias = []
    registros_df = df.shape[0]

    # A - DOCUMENTO -----------------------------------------------------------------------
    df_a = pd.DataFrame(df['DNI'].astype(int))
    df_a.rename(columns={'DNI' : 'DOCUMENTO'}, inplace=True)

    # B,C Columnas vacias-------------------------------------------------------------------
    columnas_bc = {
    'NUMEROCONVENIO' : columnasVacias,
    'NUMEROOPERACION' : columnasVacias
    }
    df_bc = pd.DataFrame(columnas_bc)

    #D - COMPAÑÍA--------------------------------------------------------------------------
    codigoCompania_df = []
    codigoCompaniaArchivo = os.path.splitext(archivoLectura)
    valorArchivo = codigoCompaniaArchivo[0][11:14]

    #print('------------------------- valor compania')
    #print(valorArchivo1)
    
    for i in range(0, registros_df):
        codigoCompania_df.append(int(valorArchivo)) 
    df_d = pd.DataFrame({'COMPANIA' : codigoCompania_df})
    
    #E - FECHAPAGO-------------------------------------------------
    listaFecha = []
    longitud = np.arange(len(df)) 

    for i in longitud:
        date_time_obj = datetime.strptime(fechaDelDia, '%Y-%m-%d %H:%M:%S') 
        date_time_obj=datetime.strftime(date_time_obj, format="%d/%m/%Y") 
        listaFecha.append(date_time_obj) 
    
    columnas_e = {
    'FECHAPAGO' : listaFecha
    }
    
    df_e = pd.DataFrame(columnas_e)
    
    #F - IMPORTECONSOLIDADO----------------------------------------------------------------------------
    df_f = pd.DataFrame(df[fechaDelDia])
    df_f.rename(columns={fechaDelDia : 'IMPORTECONSOLIDADO'}, inplace=True)
    

    # df_f = df.assign(IMPORTECONSOLIDADO = df['DEUDA'] - df['SALDO'])
    # df_a.rename(columns={'0' : 'IMPORTECONSOLIDADO'}, inplace=True)
    #df_f = pd.DataFrame('IMPORTECONSOLIDADO')

    #G, H, I, J, K Columnas vacias------------------------------------------------------------------------------------
    columnas_vacias_gk = {
        'IMPORTECAPITAL' : columnasVacias,
        'IMPORTEHONORARIOS' : columnasVacias,
        'IMPORTECOMISIONCANALPAGO' : columnasVacias,
        'OBSERVACION' : columnasVacias,
        'CANALPAGO' : columnasVacias
    }
    df_gk = pd.DataFrame(columnas_vacias_gk)

    #L - NUMERORECIBO---------------------------------------------------------------
    listaGuardarRecibos = []
    generadorRecibos = np.arange(len(df)) 
 
    for i in generadorRecibos:
        listaGuardarRecibos.append(i+1)
    
    columnaRecibo = {
        'NUMERORECIBO' : listaGuardarRecibos
    }
    df_l = pd.DataFrame(columnaRecibo) 
   
    #CONCATENAR COLUMNAS----------------------------------------------------------------
    df_concat = pd.concat([df_a, df_bc, df_d, df_e, df_f, df_gk, df_l], axis=1)
    #.drop_duplicates(subset=['DOCUMENTO', 'FECHAPAGO'], keep=False)

    #Borrar values 0 en IMPORTECONSOLIDADO
    df_concat.drop(df_concat[df_concat['IMPORTECONSOLIDADO'] < 0.0].index, inplace=True)
    #new_df = df_concat[df_concat.loc['IMPORTECONSOLIDADO']!= 0.0].drop()
    
    print(df_concat)
    
    try:
        df_concat.to_excel(escrituraArchivo, index=False)
        print("Archivo  'CredNowLoanCargar' creado correctamente")
    except:
        print("Error al crear el archivo 'CredNowLoanCargar'")
    
    


antinaEspecialLoanCargar('Archivos/1-076-AntinaEspecial-15032022test.xlsx', '2022-03-15 00:00:00', 'Archivos/generadosAntinaEspecial/testloanFormateado1234.xlsx')
```
